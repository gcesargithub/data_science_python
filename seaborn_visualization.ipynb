```python
# Seaborn SNS
# Import libraries

import seaborn as sns
import numpy as np
import pandas as pd
import sklearn
```

```python
# Set palette

sns.set_palette('deep')
sns.palplot(sns.color_palette())
```

```python
# Create DataFrame from sns dateset
# github.com/mwaskom/seaborn-data/

tips = pd.read_csv('/Users/gcesar/Documents/Code/DataScience/tips.csv')
tips.head()
```

```python
# Figure-level interface sns.relplot()
# Axes-level functions scatterplot(), lineplot()
```

```python
# Plot bill and tips scatter
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

```python
# Hue added for smoker and nonsmoker
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='smoker')
```

```python
# Size parameter
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', size='size')
```

```python
# Added segments time and gender
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='smoker', row='time', col='sex', height=3, aspect=1)
```

```python
# Hexbin jointplot(kind='hex') higher count
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex', color='black')
```

```python
# Regression line jointplot(kind='reg')
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```

```python
# Regression line with hue by gender lmplot(hue='sex', order=1)
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', order=1)
```

```python
#  Jointplot with hue parameter jointplot(hue='time') scatter is default
sns.jointplot(data=tips, x='total_bill', y='tip', hue='time')
```

```python
# Strip count is moved for visualization purposes
sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
```

```python
# Swarm count is moved for visualization purposes
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
```

```python
# Swarm with hue added for smoker and order added for day. Days with no data display blank
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm', hue='smoker', order=['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun'])
```
