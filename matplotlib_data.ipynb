```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
# Create DataFrame from Excel

df = pd.read_excel('/Users/gcesar/Desktop/DataScience/StatisticsExcelCG.xlsx')
df
```

```python
# Drop NAN
df1 = df.dropna()
```

```python
df1
```

```python
# Create X and Y
X =  df1['Job Title']
Y =  df1['# of Jobs 2026 Est.']
```

```python
# Plot bar chart
plt.bar(x=X, height=Y, width=0.5)
plt.xlabel('Jobs')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=90)
plt.title('Number of Tech Jobs 2026')
```

```python
# Pie plot
plt.pie(Y, labels=X)
```

```python
# Stacked bar plot
Y1 =  df1['# of Jobs 2016']
plt.bar(x=X, height=Y, bottom=None)
plt.bar(x=X, height=Y1, bottom=Y)
plt.xlabel('Jobs')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=90)
plt.title('Number of Tech Jobs 2016, 2026')
```

```python
# Subplots
plt.subplot(2, 2, 1)
plt.bar(x=X, height=Y1)
plt.xlabel('Jobs')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=90)
plt.title('Number of Tech Jobs 2016')

plt.subplot(2, 2, 2)
plt.bar(x=X, height=Y)
plt.xlabel('Jobs')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=90)
plt.title('Number of Tech Jobs 2026')
```
